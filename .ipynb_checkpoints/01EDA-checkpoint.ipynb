```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
df = pd.read_csv("data/UCI_Credit_Card.csv")
df.columns = df.columns.str.strip()  
df.head()
```

```python
# Shape and column info
print("Shape:", df.shape)
print("\nColumn types:\n", df.dtypes)

# Quick summary statistics
df.describe().T
```

```python
target = "default.payment.next.month"
df[target].value_counts(normalize=True).plot(kind="bar", rot=0, color=["skyblue","salmon"])
plt.title("Default vs Non-default Distribution")
plt.ylabel("Proportion")
plt.show()
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True)[[target]].sort_values(by=target, ascending=False), annot=True, cmap="coolwarm")
plt.title("Correlation of Features with Target")
plt.show()
```

```python
num_cols = ["LIMIT_BAL", "AGE", "PAY_0"]  # example important columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=False)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
df_clean = df.drop(columns=["ID"], errors="ignore")
df_clean.to_csv("data/credit_clean.csv", index=False)
df_clean.shape
```
